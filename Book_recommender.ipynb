# Popularity Based 

```python
ratings_new=ratings.merge(books, on="ISBN")
```

```python
num_rating_df=ratings_new.groupby("Book-Title").count()["Book-Rating"].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
# Convert 'Book-Rating' column to numeric, setting errors='coerce' will convert non-numeric values to NaN
ratings_new["Book-Rating"] = pd.to_numeric(ratings_new["Book-Rating"], errors="coerce")

# Drop rows with NaN values in 'Book-Rating' column
ratings_new = ratings_new.dropna(subset=["Book-Rating"])

# Group by 'Book-Title' and calculate the mean rating
avg_rating_df = ratings_new.groupby("Book-Title").mean(numeric_only=True)["Book-Rating"].reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df["num_ratings"]>=250].sort_values("avg_rating", ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books , on="Book-Title").drop_duplicates("Book-Title")[["Book-Title", "Book-Author","Image-URL-M","num_ratings", "avg_rating"]]
```

```python
popular_df["Image-URL-M"][0]
```

```python
pickle.dump(popular_df ,open("popular.pkl","wb"))
```

# Collabrative Filtering System

```python
ratings_new
```

```python
x= ratings_new.groupby("User-ID").count()["Book-Rating"] > 200
```

```python
more_than_two=x[x].index
```

```python
filtered_data=ratings_new[ratings_new["User-ID"].isin(more_than_two)]
```

```python
y= filtered_data.groupby("Book-Title").count()["Book-Rating"]>=50
```

```python
y[y].index
```

```python
more_than_fifty = y[y].index
```

```python
filtered_ratings=filtered_data[filtered_data["Book-Title"].isin(more_than_fifty)]
```

```python
filtered_ratings
```

```python
pt=filtered_ratings.pivot_table(index="Book-Title" , columns="User-ID", values="Book-Rating") 
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt).shape
```

```python
Similarity = cosine_similarity(pt)
```

```python
def Recommendation(Book_name):
    index=np.where(pt.index==Book_name)[0][0]
    Similar_items = sorted(list(enumerate(Similarity[index])),key= lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in Similar_items: 
        item=[]
        temp_df = books[books["Book-Title"] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        data.append(item)
    return data
```
